```python
from engine.fields.mana_field import ManaField
from engine.math.b_calculus import b_derivative

mana = ManaField(shape=(10, 10), initial_value=1.0)
print("Initial mean mana:", mana.grid.mean())

# apply B-scale power 2 (like ⊗ 2)
mana.b_scale_power(2.0)
print("After power 2, mean mana:", mana.grid.mean())

# test a simple B-derivative numerically on f(x) = x**3
f = lambda x: x**3
for x in [1.0, 2.0, 4.0]:
    print("D_B f at x =", x, "≈", b_derivative(f, x))
    # (should be close to e^3 ≈ 20.085)

mana = ManaField(shape=(10,10), initial_value=2.0)
mana.b_scale_power(3.0)
print(mana.grid.mean())
```
